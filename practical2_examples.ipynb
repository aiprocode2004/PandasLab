```python
#Program 1: Student Performance Analysis 
import pandas as pd 
# Create Dataset 
data = { 
"Student":["Amit","Rahul","Priya","Neha","Karan"], 
"Math":[85,76,92,65,88], 
"Science":[90,81,95,70,86], 
"English":[78,85,89,75,80] 
} 
df = pd.DataFrame(data) 
print(df) 
# Total Marks 
df["Total"] = df["Math"] + df["Science"] + df["English"] 
# Percentage 
df["Percentage"] = df["Total"]/3 
print("\nStudent Report") 
print(df) 
print("\nTop Performer") 
print(df.loc[df["Percentage"].idxmax()]) 
print("\nAverage Percentage") 
print(df["Percentage"].mean())
```

```python
#Program 2: Employee Salary Analysis 
import pandas as pd 
employees = { 
"Employee":["John","David","Maria","Sara","Alex"], 
"Department":["HR","IT","IT","Finance","HR"], 
"Salary":[35000,65000,72000,55000,42000] 
} 
df = pd.DataFrame(employees) 
print(df) 
print("\nAverage Salary") 
print(df["Salary"].mean()) 
print("\nHighest Salary") 
print(df[df["Salary"]==df["Salary"].max()]) 
print("\nDepartment Wise Salary") 
print(df.groupby("Department")["Salary"].mean()) 
```

```python
#Program 3: Sales Data Analytics 
import pandas as pd 
sales = { 
"Product":["Laptop","Mobile","Laptop","Tablet","Mobile"], 
"Quantity":[5,10,7,4,6], 
"Price":[50000,20000,50000,30000,20000] 
} 
df = pd.DataFrame(sales) 
df["Revenue"] = df["Quantity"] * df["Price"] 
print(df) 
print("\nTotal Revenue") 
print(df["Revenue"].sum()) 
print("\nRevenue by Product") 
print(df.groupby("Product")["Revenue"].sum()) 
```

```python
#Program 4: Hospital Patient Analytics 
import pandas as pd 
patients = { 
"Patient":["P1","P2","P3","P4","P5"], 
"Age":[25,40,32,67,55], 
"Disease":["Diabetes","Fever","Diabetes","Cancer","Fever"] 
} 
df = pd.DataFrame(patients) 
print(df) 
print("\nDisease Count") 
print(df["Disease"].value_counts()) 
print("\nAverage Age") 
print(df["Age"].mean()) 
```

```python
#Program 5: Banking Transaction Analysis 
import pandas as pd 
transactions = { 
"Customer":["A","B","C","A","B"], 
"Amount":[5000,-2000,10000,-1500,7000] 
} 
df = pd.DataFrame(transactions) 
print(df) 
print("\nTotal Deposits") 
print(df[df["Amount"]>0]["Amount"].sum()) 
print("\nTotal Withdrawals") 
print(df[df["Amount"]<0]["Amount"].sum())
```

```python
#Program 6: Online Shopping Analytics 
import pandas as pd 
orders = { 
"OrderID":[101,102,103,104], 
"Customer":["Ram","Shyam","Hari","Mohan"], 
"Amount":[2500,3200,1800,4100] 
} 
df = pd.DataFrame(orders) 
print(df) 
print("\nHighest Order") 
print(df.loc[df["Amount"].idxmax()]) 
print("\nAverage Order Value") 
print(df["Amount"].mean()) 
```
